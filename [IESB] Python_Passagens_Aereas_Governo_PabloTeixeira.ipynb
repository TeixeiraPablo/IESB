```python
#importação de bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cinco Órgãos superiores com maior quantidade de passagens emitidas
df['Nome do órgão superior'].value_counts().head()
```

```python
# Cinco Órgãos com maior quantidade de solicitações 
df['Nome do órgão solicitante da viagem'] = df['Nome do órgão solicitante da viagem'].str.lstrip()
df['Nome do órgão solicitante da viagem'] = df['Nome do órgão solicitante da viagem'].str.rstrip()
df['Nome do órgão solicitante da viagem'].value_counts().head()
```

```python
# Passagens por companhia aérea
df['Companhia Aérea'].value_counts()
```

```python
# Passagens por situação
df['Situação Final Bilhete'].value_counts(normalize=True) # informação em percentual
```

```python
# Soma dos valores gastos
df['Valor Bilhete'].sum()
```

```python
# Bilhete mais barato
df['Valor Bilhete'].min()
```

```python
# Bilhete mais caro
df['Valor Bilhete'].max()
```

```python
# Preço médio do bilhete
df['Valor Bilhete'].mean()
```

```python
# Quantidade de bilhetes emitidos por companhia aerea
df.groupby('Companhia Aérea')['Valor Bilhete'].count()
```

```python
# Soma dos valores gastos por companhia áerea
df.groupby('Companhia Aérea')['Valor Bilhete'].sum()
```

```python
# Soma, média, total e desvio padrão dos valores gastos por companhia aerea
df.groupby('Companhia Aérea')['Valor Bilhete'].agg(['sum', 'mean', 'count','std'])
```

```python
# Redimensionar o gráfico
sns.set(rc={'figure.figsize': (16.7,8.27)})
```

```python

df['Companhia Aérea'] = df['Companhia Aérea'].str.lstrip()
df['Companhia Aérea'] = df['Companhia Aérea'].str.rstrip()
```

```python
# Gráfico de distribuição normal - Preço do bilhete por companhia
sns.distplot(df[df['Companhia Aérea'] == 'AVIANCA S.A.']['Valor Bilhete'].dropna(),label='AVIANCA')
sns.distplot(df[df['Companhia Aérea'] == 'AZUL LINHAS AÉREAS BRASILEIRA']['Valor Bilhete'].dropna(),label='AZUL')
sns.distplot(df[df['Companhia Aérea'] == 'GOL Linhas Aéreas Inteligentes']['Valor Bilhete'].dropna(),label='GOL')
sns.distplot(df[df['Companhia Aérea'] == 'LATAM AIRLINES BRASIL']['Valor Bilhete'].dropna(),label='LATAM')
sns.distplot(df[df['Companhia Aérea'] == 'MAP TRANSPORTES AEREOS LTDA - MAP']['Valor Bilhete'].dropna(),label='MAP')
sns.distplot(df[df['Companhia Aérea'] == 'Gol OU Latam - Parceria com Terceiro']['Valor Bilhete'].dropna(),label='PARCERIA GOL LATAM')
plt.legend()
```

```python
# boxplot - valor da passagem por companhia
sns.boxplot(x='Companhia Aérea', y='Valor Bilhete', data=df)
```

```python
# Situação dos voos
sns.countplot(x='Situação Final Bilhete', data=df)
```

```python
# violin plot
sns.violinplot(x='Companhia Aérea', y='Valor Bilhete', data=df)
```

```python
# Voos "no show"
sns.countplot(x='No Show', data=df)
```

```python
# Voos "cancelados"
sns.set_palette('Set2')
sns.countplot(x='Cancelado', data=df)
```

```python
# Situacao voo por companhia
sns.factorplot(x='Situação Final Bilhete', data=df, col='Companhia Aérea', kind='count')
```

```python
# Valor do bilhete por companhia
sns.factorplot(x='Situação Final Bilhete', y='Valor Bilhete', data=df, col='Companhia Aérea', kind='violin')
```
